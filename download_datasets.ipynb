```python
from matplotlib import pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(8, 4))

ax.grid("--", color="gray", alpha=0.4)

ax.plot(np.random.rand(128), color="royalblue", label="Train Loss")
ax.plot(np.random.rand(128), color="darkorange", label="Val Loss")

ax.legend(loc="upper right", fontsize=10.5)
ax.set_xlabel("Epochs", fontsize=12)
ax.set_ylabel("Classification Loss", fontsize=12)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

from seaborn import heatmap

confusion = heatmap(
    np.random.rand(10, 10) * 100,
    square=True,
    annot=True,
    # fmt="d",
    cmap="Blues",
    cbar=False,
    ax=ax,
    vmin=0,
    linecolor="k",
    linewidths=0.4,
    annot_kws={"size": 10},
)
cbar = confusion.figure.colorbar(
    confusion.collections[0], ax=confusion, location="right", shrink=0.8
)
ax.set_xlabel("Predicted label", fontsize=13)
ax.set_ylabel("True label", fontsize=13)
```
